```python
import pandas as pd                                     # importing Library
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
from wordcloud import ImageColorGenerator
from wordcloud import STOPWORDS
```

```python
df=pd.read_csv('zomato.csv')     # reading zomato files
```

```python
df.info() #  file details
```

```python
df.shape      #   fetching file size
```

```python
df     # df is the variable name of the file name
```

```python
df.drop(["url","address","phone","reviews_list","menu_item"],axis=1,inplace=True) # dropping
df.info()
```

```python
df.head() # fetching  fisrt 5 row
```

```python
pd.isnull(df).sum() # checking null value
```

```python
df["location"]=df["location"].fillna("Location not available") # renaming columns
df["rest_type"]=df["rest_type"].fillna("No rest_type")
df["dish_liked"]=df["dish_liked"].fillna("Not available")
df["cuisines"]=df["cuisines"].fillna("Not Mentioned")
```

```python
df["rate"].unique()    # fetching unique values
```

```python
def replacing(data):                                  
    if data=="NEW" or data=="-":
        return np.nan
    else:
        data=str(data).split("/")                          #replacing ("NEW" and "-" ) to Nan value
        data=data[0]
        return float(data)
df['rate']=df['rate'].apply(replacing)   
```

```python
df["approx_cost(for two people)"]=df["approx_cost(for two people)"].fillna("0")
df["approx_cost(for two people)"]=df["approx_cost(for two people)"].astype(str)      # replacing and converting to float
df["approx_cost(for two people)"]=df["approx_cost(for two people)"].apply(lambda i:i.replace(",",""))     
df["approx_cost(for two people)"]=df["approx_cost(for two people)"].astype(float)
df["approx_cost(for two people)"].unique()
```

```python
df.isnull().sum()     # fetching null  values 
```

```python
df.rename(columns={"approx_cost(for two people)":"Cost","listed_in(type)":"Type_of_Resturants","listed_in(city)":"Cities"},inplace=True)
df.info()
```
